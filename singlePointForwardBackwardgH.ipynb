```python
import numpy as np

#Define activation, activation derivative f',loss
def sigmoid(x):
    s=1/(1+np.exp(-x))
    return s

def sigmoid_derivative(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)
    return ds

def loss(x,y):
    return 0.5*(y-x)**2


#Taking arbitrary values for weights, bias, and input layer

#3x2
W_layer_1 = np.array([[ 0.15,0.45],
                     [-0.35,-0.61],
                     [-0.52, 0.22]])


#1x3
W_layer_2 = np.array([[-0.72,1.22,-0.3]])

#3x1
B_layer_1 = np.array([[0.4],
                      [0.65],
                      [0.23]])

B_layer_2 = 0.15

X = np.array([[0.1],
              [0.35]])

y=0.25

eta = 0.4

#Forward-----------------

z_layer_1 = np.dot(W_layer_1,X) + B_layer_1
a_layer_1 = sigmoid(z_layer_1)

z_layer_2 = np.dot(W_layer_2,a_layer_1) + B_layer_2
a_layer_2 = sigmoid(z_layer_2)

print('a4',a_layer_2)

#Backward--------------------------


#deltas -- note * is shur product
delta2 = (2*(a_layer_2-y))*sigmoid_derivative(z_layer_2)

#3x1  1x1 -> 3x1
delta1 = np.dot(W_layer_2.T,delta2) * sigmoid_derivative(z_layer_1)

# 1x2 3x1 -> 3x1 1x2 - > 3x2
dlossdW1 = np.dot(delta1,X.T)

#deltaLoss/deltaw2 = (al1)*f'(zL)(-(y-a4))

dlossdW2 = np.dot(delta2,a_layer_1.T)


#update weights
W_layer_1 = W_layer_1 - eta*dlossdW1

W_layer_2 = W_layer_2 - eta*dlossdW2

#update Bias
#we know deltaL/delta B = delta_L
#1x3 - 1x3 (Transpose)
B_layer_1 = B_layer_1 - eta*delta1

B_layer_2 = B_layer_2 - eta*delta2

print('\nW_layer_1 updated \n',W_layer_1,(-1,1))

print('\n\nW_layer_2 updated \n',W_layer_2,(-1,1))

print('\nB_layer_1 updated \n',B_layer_1,(-1,1))

print('\n\nB_layer_2 updated \n',B_layer_2,(-1,1))
```
